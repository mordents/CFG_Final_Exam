```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe: 
netflix = pd.read_csv('netflix_titles.csv')
netflix
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.
netflix2=netflix.dropna(how='any')      #no missing values - no nan #
rows=netflix.shape[0]-netflix2.shape[0]     # no of rows in og frame - no of rows in clean frame#
rows
```

```python
netflix.dropna(how='any',inplace=True )         # dropping rows here bc earlier I didn't do it properly#
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame.
netflix['type']=netflix['type'][netflix['type']=='Movie']
netflix
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average 
  # duration of the movies in the dataset.
netflix.dropna(how='any',inplace=True )  

netflix['duration']=netflix['duration'].str.replace("min","")
netflix['duration']=netflix['duration'].astype(float)
avgdur=netflix['duration'].mean()
```

```python
netflix['duration']=netflix['duration'].fillna(avgdur)
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
netflix.drop_duplicates(inplace=True) 
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter. 
netflix.columns=netflix.columns.str.title() #using python method#
netflix #checx if it worked
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file. 
netflix.to_csv('cleaned_netflix_data.csv')
```

2. MATPLOTLIB (25 points)

Use the netflix dataset to solve the exercises in this section.

```python
# 1 (10 marks). Create a visualisation which shows the frequency of movies by year of release.
netflix.groupby('Release_Year')['Type'].count().plot(kind='bar', figsize=(20,20),fontsize=15)
fontsize=20
plt.ylabel('Count',fontsize=20)
plt.xlabel('Release_Year',fontsize=20)
plt.title('Frequency of movies by year', fontsize=20)
plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movies 
  # in the dataset.
netflix.groupby('Listed_In')['Type'].count().sort_values(ascending=False).head(5).plot.pie()
plt.show()
```
